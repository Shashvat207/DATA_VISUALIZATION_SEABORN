```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.style.use("fivethirtyeight")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

# Catgorical scatter plots , Striplots and Swarmplots

```python
sns.catplot(x="day",y="tip",kind="strip",data=tips)
```

```python
# jitter parameter makes the same values have noise and make them visualize as non-overlapping
#changing the value of jitter can induce more noice
sns.catplot(x="day",y="tip",kind="strip",data=tips,jitter=1.5)
plt.show()
sns.catplot(x="day",y="tip",kind="strip",data=tips,jitter=0)
plt.show()
```

### SWARMPLOTS

```python
#swarm plots also show distribution of data 
sns.catplot(x="day",y="tip",kind="swarm",data=tips)
plt.show()
```

```python
sns.catplot(x="day",y="tip",kind="strip",hue="sex",data=tips)
```

# BOXPLOTS

```python
#univariate
sns.boxplot(tips.total_bill)
```

```python
#swarm plots also show distribution of data 
sns.catplot(x="day",y="total_bill",kind="box",data=tips)
plt.show()
```

```python
#swarm plots also show distribution of data 
sns.catplot(x="day",y="total_bill",kind="box",hue="sex",data=tips)
plt.show()
```

# VIOLIN PLOTS

```python
sns.violinplot(tips.total_bill)
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=tips)
plt.show()
```

```python
#swarm plots also show distribution of data 
sns.catplot(x="day",y="total_bill",kind="violin",hue="sex",data=tips)
plt.show()
```

```python
#swarm plots also show distribution of data 
sns.catplot(x="day",y="total_bill",kind="violin",hue="sex",split=True,data=tips)
plt.show()
```

# BARPLOTS AND COUNTPLOTS

```python
sns.catplot(x="smoker",y="total_bill",kind="bar",data=tips)
```

```python
sns.catplot(x="smoker",y="total_bill",kind="bar",data=tips,hue="sex")
```

```python
#estimator can be np.mean,np.median,np.std,np.var depending on different requirements
sns.catplot(x="smoker",y="total_bill",kind="bar",data=tips,hue="sex",estimator=np.mean)
```

```python
#the way histogram works on numerical data count plort works on categorical data
sns.catplot(x="smoker",data=tips,kind="count")
```

# HEATMAPS

```python
flights=sns.load_dataset("flights")
```

```python
flights.head()
x=flights.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,annot=True,fmt="d",linewidths=0.5)
```

```python
#cmap can be viridis , summer  ,winter and more
sns.heatmap(x,annot=True,fmt="d",linewidths=0.5,cmap="winter")
```

# CLUSTERMAPS

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric="correlation")
```

```python
sns.clustermap(x,metric="correlation",z_score=0,annot=True)
```

# DISTPLOTS

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic["age"].fillna(titanic["age"].mean(),inplace=True)
```

```python
sns.distplot(titanic["age"])
```

```python
sns.distplot(titanic["age"],kde=False)
```

```python
sns.distplot(titanic["age"],hist=False)
```

```python
sns.distplot(titanic["age"],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic["survived"]==1]["age"])
sns.distplot(titanic[titanic["survived"]==0]["age"])
```
